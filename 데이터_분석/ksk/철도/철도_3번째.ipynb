```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train2019 = pd.read_csv('2019철도.csv', thousands=',', encoding='utf-8')
train2020 = pd.read_csv('2020철도.csv', thousands=',', encoding='utf-8')
train2021 = pd.read_csv('2021철도.csv', thousands=',', encoding='utf-8')
```

```python
train2019['월'] = train2019['월별'].str[4:]
train2019['년'] = train2019['월별'].str[:4]
```

```python
train2020['월'] = train2020['월별'].str[4:]
train2020['년'] = train2020['월별'].str[:4]
```

```python
train2021['월'] = train2021['월별'].str[4:]
train2021['년'] = train2021['월별'].str[:4]
```

```python
a = train2019.groupby('월')['총 이용객수'].sum()
b = train2020.groupby('월')['총 이용객수'].sum()
c = train2021.groupby('월')['총 이용객수'].sum()
```

```python
# 한글 문제
# matplotlit의 기본 폰트에서 한글 지원되지 않기 때문에
# matplotlib의 폰트 변경 필요
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
plt.figure(figsize=(10, 10))
plt.plot(a.index, a,label='2019')
plt.bar(b.index, b,align='center',color='green',label='2020',width=0.35)
plt.bar(c.index, c,width=0.35,align='edge',color='purple',label='2021')
plt.legend()
plt.xlabel('월')
plt.ylabel('철도 이용객수')
plt.title('3년간 철도 고객 이용객 변화')
plt.show()
plt.savefig('철도이용객.png',dpi=100)
```

```python
total_train = pd.concat([train2019,train2020,train2021], ignore_index=True)
total_train
```

```python
total_train1= total_train.sort_values(by = ['월', '총 이용객수'])
sns.catplot(data = total_train1, x = '월', y = '총 이용객수', hue = '년', kind = 'bar', 
            palette = 'Reds', legend_out = True)
```
