### 1.이전에 작업한 2020데이터(01~07) 가져오기

```python
os.chdir(path = '../result')
curpath=os.getcwd()
curpath

flight_2020=pd.read_csv('../result/flight_2020.csv')
flight_2020
```

```python
flight_dom_2020=[] #국내운항편
flight_intl_2020=[] #국제운항편
psger_dom_2020=[] #국내여객자수
psger_intl_2020=[]  #국외여객자수

flight_dom_2019=[]
flight_intl_2019=[]
psger_dom_2019=[]
psger_intl_2019=[]
```

```python
for flight_dom in flight_2020['국내 운항편']:
    flight_dom_2020.append(flight_dom)

for flight_intl in flight_2020['국제 운항편']:
    flight_intl_2020.append(flight_intl)

for psger_dom in flight_2020['국내 여객자수']:
    psger_dom_2020.append(psger_dom)

for psger_intl in flight_2020['국제 여객자수']:
     psger_intl_2020.append(psger_intl)
```

```python
#2020.01~ 2020.07 기존data 
len(flight_dom_2020)
flight_dom_2020
flight_intl_2020
psger_dom_2020
psger_intl_2020
```
