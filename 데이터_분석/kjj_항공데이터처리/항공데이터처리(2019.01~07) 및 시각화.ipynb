### 2. 시각화

```python
flight_2020=pd.read_csv('./result/flight_2020.csv')
flight_2020.rename(columns={'Unnamed: 0':'날짜'},inplace=True)
flight_2020

flight_2021=pd.read_csv('./result/flight_2021.csv')
flight_2021.rename(columns={'Unnamed: 0':'날짜'},inplace=True)
flight_2021
```

### 2020, 2021 데이터합치기

```python
flight2=pd.concat([flight_2020,flight_2021])
flight2
```

```python
flight2['날짜'] = pd.to_datetime(flight2['날짜'])
flight2['년'] = flight2['날짜'].dt.year
flight2['월'] = flight2['날짜'].dt.month

flight2['총 운항편'] = flight2['국내 운항편'] +  flight2['국제 운항편']
flight2['총 여객자수'] = flight2['국내 여객자수'] +  flight2['국제 여객자수']
flight2
```

```python
# 2019데이터 
flight_2019=pd.read_csv('./result/flight_2019.csv')
flight_2019.rename(columns={'Unnamed: 0':'날짜'},inplace=True)
flight_2019

flight_2019['날짜'] = pd.to_datetime(flight_2019['날짜'])
flight_2019['년'] = flight_2019['날짜'].dt.year
flight_2019['월'] = flight_2019['날짜'].dt.month
#flight_2019[['날짜','년','월']]

flight_2019['총 운항편'] = flight_2019['국내 운항편'] +  flight_2019['국제 운항편']
flight_2019['총 여객자수'] = flight_2019['국내 여객자수'] +  flight_2019['국제 여객자수']
flight_2019
```

```python
flight3=pd.concat([flight_2019,flight])
flight3
```
